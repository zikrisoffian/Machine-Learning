## Machine Learning Basic Algorithm : Cost Function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# sample data

x = np.array([1,2,3,4,5])
y = np.array([2,3.5,4.5,5.5,6.5])
```

```python
plt.scatter(x,y)
```

```python
# hyperparameter

learning_rate = 0.001
epochs = 1000
```

```python
# initial parameter (coeffiecient and y-intercept)

slope = 0
intercept = 0

# store variable values in list

slope_values = []
intercept_values = []
mse_values = []
```

```python
for i in range(epochs):
    y_pred = slope * x + intercept
    error = y_pred - y
    mse = np.mean(error ** 2)
    
    # Gradient Descent Algorithm
    slope_gradient = 2 * np.dot(error, x) / len(x)
    intercept_gradient = 2 * np.sum(error) / len(x)
    slope -= learning_rate * slope_gradient
    intercept -= learning_rate * intercept_gradient
    
    # Store values in lists
    
    slope_values.append(slope)
    intercept_values.append(intercept)
    mse_values.append(mse)
```

```python
epoch_value = np.arange(1,1000+1)
```

```python
# slope_values visualization

plt.plot(epoch_value,slope_values)
```

```python
# intercept_values visualization

plt.plot(epoch_value,intercept_values)
```

```python
# Plot the progress of gradient descent
plt.figure(figsize = (12,8))

# Plot the change in the slope (coefficient)

plt.subplot(1,2,1)
plt.plot(slope_values, mse_values)
plt.xlabel('Coefficient (slope)')
plt.ylabel('Mean Squared Error')

# Plot the change in the y-intercept (bias)

plt.subplot(1,2,2)
plt.plot(intercept_values, mse_values)
plt.xlabel('Coefficient (slope)')
plt.ylabel('Mean Squared Error')

plt.show()

print("Slope: ", slope)
print("Intercept: ", intercept)
```
